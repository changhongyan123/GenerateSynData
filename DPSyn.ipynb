```python
import csv
import pandas as pd
import random
from collections import OrderedDict
```

```python
data = pd.read_csv('dataset', header=None)
```

```python
data.shape
```

```python
L = 8
```

```python
def sub_table(L, view):
    m = data.shape[0] # or smaller number 
    sub_origin_table = data[data.columns[view]].sample( n = m ) # select the view attributes from original dataset
    return sub_origin_table
```

```python
# generate all the permutation of 0,1 in size m  (string)
permutation_number = (1<<L)-1
permutations = []
for i in range(permutation_number+1):
    fo = '0' + str(L) + 'b' # format length
    number = format(i, fo ) # change to binary concat
    permutations.append(number)
    
print(permutations)
```

```python
total_view = []

def generating_counting_matrix(L):
    # random select attribute as a view
    attributes = range(data.shape[1]) # All the attributes in the dataset
    view = random.sample(attributes,L) # randomly select m attributes
    view = sorted(view)
#     print(view)
    total_view.append(view)
    
    # sub_sampling from the original data
    sub_origin_table = sub_table(L, view)
    mapping_row_to_binary = sub_origin_table.T.to_dict('list')
    
   

    dictOfCount = { i:0 for i in permutations}
    for x,y in mapping_row_to_binary.items():
        binary_to_string = ''.join(map(str, y)) # change v in k,v to string of binary
        dictOfCount[binary_to_string] = dictOfCount.get(binary_to_string,0) + 1 # update count
    return pd.DataFrame(list(dictOfCount.items()), columns = ['permu',view]).T.iloc[[1]]
```

```python
result = generating_counting_matrix(L)
```

```python
for i in range(2):
    sample = generating_counting_matrix(L)
    result = result.append(sample)
```
